# seaborn_altair.catplot

```python
import numpy as np
import seaborn as sns
import seaborn_altair as salt
exercise = sns.load_dataset("exercise")
```

```python
display(salt.catplot(x="time", y="pulse", hue="kind", data=exercise, kind="bar"))
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise, kind="bar")
```

```python
display(salt.catplot(x="time", y="pulse", hue="kind", data=exercise, kind="bar"))
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise, kind="violin")
```

```python
display(salt.catplot(x="kind", y="pulse", col="time", row="diet", kind="bar", data=exercise))
g = sns.catplot(x="time", y="pulse", hue="kind", col="diet", kind="bar", data=exercise)
```

```python
display(salt.catplot(x="time", y="pulse", row="kind", col="diet", kind="bar", data=exercise))
g = sns.catplot(x="time", y="pulse", row="kind", col="diet", kind="bar", data=exercise)
```

```python
titanic = sns.load_dataset("titanic")
display(salt.catplot("alive", col="deck",
                     data=titanic[titanic.deck.notnull()],
                     kind="count", height=2.5, aspect=.8))
g = sns.catplot("alive", col="deck", col_wrap=4,
                data=titanic[titanic.deck.notnull()],
                kind="count", height=2.5, aspect=.8)
```

```python
display(salt.catplot(x="age", y="embark_town",
                     hue="sex", row="class",
                     data=titanic[titanic.embark_town.notnull()],
                     orient="h", height=2, aspect=3.5, palette="Set3",
                     kind="strip"))
g = sns.catplot(x="age", y="embark_town",
                hue="sex", row="class",
                data=titanic[titanic.embark_town.notnull()],
                orient="h", height=2, aspect=3.5, palette="Set3",
                kind="violin", dodge=True, cut=0, bw=.2)
```
