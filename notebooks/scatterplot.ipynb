```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import seaborn_altair as salt
tips = sns.load_dataset("tips")
```

```python
display(salt.scatterplot(x="total_bill", y="tip", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", hue="time", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", size="size", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
display(salt.scatterplot(x="total_bill", y="tip", hue="size", size="size", palette=cmap, data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", palette=cmap, data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
display(salt.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 100), palette=cmap, data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20, 100), palette=cmap, legend="full", data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", hue="day", size="smoker", sizes=[10, 80], palette="Set2", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", hue="day", size="smoker", palette="Set2", data=tips)
```

```python
markers = {"Lunch": "s", "Dinner": "X"}
display(salt.scatterplot(x="total_bill", y="tip", style="time", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", style="time", markers=markers, data=tips)
```

```python
display(salt.scatterplot(x="total_bill", y="tip", data=tips))
ax = sns.scatterplot(x="total_bill", y="tip", s=100, color=".2", marker="+", data=tips)
```

```python
iris = sns.load_dataset("iris")
display(salt.scatterplot(x=iris.sepal_length, y=iris.sepal_width, hue=iris.species, style=iris.species))
ax = sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width, hue=iris.species, style=iris.species)
```

```python
import numpy as np, pandas as pd; plt.close("all")
index = pd.date_range("1 1 2000", periods=100, freq="m", name="date")
data = np.random.randn(100, 4).cumsum(axis=0)
wide_df = pd.DataFrame(data, index, ["a", "b", "c", "d"])
display(salt.scatterplot(data=wide_df))
ax = sns.scatterplot(data=wide_df)
```
